```python
task['fare_group']=pd.cut(task['Fare'],bins=[0,10,30,100],labels=['low','medium','high'])
task['fare_group']
```

```python
task['Age_group']=pd.cut(task['Age'],bins=[0,12,18,40,60,100],labels=['child','teenager','youth','adult','senior'])
task['Age_group']
```

```python
task["title"]=task['Name'].apply(lambda x: x.split(',')[1].split(',')[0].strip())
task["title"]
```

```python
task
```

```python
sns.countplot(x='fare_group', hue='Survived',data=task)
plt.title('survival count by Fare class')
plt.xlabel('Fare class')
plt.show()
```

From the graph above, we discover that the highest people that did not survived were those that payed a low fare followed by those that paid medium fare and finally those that paid high fare.


```python
sns.countplot(x='Age_group', hue='Survived',data=task)
plt.title('survival count by Age_group class')
plt.xlabel('Age class')
plt.show()
```

From the graph above we discover that the number of people that survive in the child class where more than those that died. but when approaching teenager class you will see that the number of people that died is already overshadowing the number that survived.the trend continues till arriving senior class you will find out that only a few people survived

```python
sns.countplot(x='title', hue='Survived',data=task)
plt.title('survival count by title class')
plt.xlabel('title class')
plt.show()
```

```python
sns.countplot(x='Pclass', hue='Fare',data=task)
plt.title('Fare count by Pclass')
plt.xlabel(' Pclass')
plt.show()
```

```python
sns.countplot(x='family_size', hue='Survived',data=task)
plt.title('survival count by Family_size class')
plt.xlabel('Family class')
plt.show()
```

From the graph above, We see that those with only one family member were likely to survived as compare to those with more family members. That's why arriving at those with family class of eleven, we finds out that no body survived or only a few that are invisible on the graph survived.
